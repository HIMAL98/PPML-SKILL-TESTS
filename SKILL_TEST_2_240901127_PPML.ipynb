## Mathematical Calculations: Euclidean Distance and Dot Product

### 1. Calculate Euclidean Distance for Two Points in 2D Space

```python
import math

def euclidean_distance(point1, point2):
    """
    Calculates the Euclidean distance between two points in 2D space.

    Args:
        point1 (tuple): A tuple representing the first point (x1, y1).
        point2 (tuple): A tuple representing the second point (x2, y2).

    Returns:
        float: The Euclidean distance between the two points.
    """
    x1, y1 = point1
    x2, y2 = point2
    distance = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)
    return distance

# Example usage:
p1 = (1, 1)
p2 = (4, 5)
dist = euclidean_distance(p1, p2)
print(f"The Euclidean distance between {p1} and {p2} is: {dist}")

p3 = (0, 0)
p4 = (3, 4)
dist2 = euclidean_distance(p3, p4)
print(f"The Euclidean distance between {p3} and {p4} is: {dist2}")
```

```python
def dot_product_matrices(matrix_a, matrix_b):
    """
    Calculates the dot product (matrix multiplication) of two matrices.

    Args:
        matrix_a (list of lists): The first matrix.
        matrix_b (list of lists): The second matrix.

    Returns:
        list of lists or str: The resulting matrix or an error message if dimensions are incompatible.
    """
    # Get dimensions of matrices
    rows_a = len(matrix_a)
    cols_a = len(matrix_a[0])
    rows_b = len(matrix_b)
    cols_b = len(matrix_b[0])

    # Check for compatible dimensions for dot product
    if cols_a != rows_b:
        return "Error: Number of columns in the first matrix must be equal to the number of rows in the second matrix."

    # Initialize the result matrix with zeros
    result_matrix = [[0 for _ in range(cols_b)] for _ in range(rows_a)]

    # Perform matrix multiplication
    for i in range(rows_a): # Iterate through rows of matrix_a
        for j in range(cols_b): # Iterate through columns of matrix_b
            for k in range(cols_a): # Iterate through columns of matrix_a (or rows of matrix_b)
                result_matrix[i][j] += matrix_a[i][k] * matrix_b[k][j]
    return result_matrix

# Example usage:
matrix1 = [
    [1, 2, 3],
    [4, 5, 6]
]

matrix2 = [
    [7, 8],
    [9, 1],
    [2, 3]
]

result = dot_product_matrices(matrix1, matrix2)

print("Matrix A:")
for row in matrix1:
    print(row)

print("\nMatrix B:")
for row in matrix2:
    print(row)

print("\nDot Product (A @ B):")
if isinstance(result, str):
    print(result)
else:
    for row in result:
        print(row)


# Another example with incompatible matrices
matrix3 = [
    [1, 2],
    [3, 4]
]

matrix4 = [
    [5, 6, 7],
    [8, 9, 0]
]

result2 = dot_product_matrices(matrix4, matrix3)
print("\nDot Product (B @ A - incompatible dimensions):")
if isinstance(result2, str):
    print(result2)
else:
    for row in result2:
        print(row)
```

## Time Series Data Analysis and Visualization

### 1. Generate a dummy time series dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Set a random seed for reproducibility
np.random.seed(42)

# Define the date range for one year
dates = pd.date_range(start='2023-01-01', periods=365, freq='D')

# Generate dummy daily temperatures
# Base temperature around 15-20 degrees Celsius
# Add some seasonality (e.g., sine wave for temperature fluctuations over a year)
# Add some random noise

base_temp = 15
seasonality = 10 * np.sin(np.linspace(0, 2 * np.pi, 365)) # Simulating yearly temperature cycle
noise = np.random.normal(0, 2, 365) # Random daily fluctuations

daily_temperatures = base_temp + seasonality + noise

# Create a Pandas DataFrame
df_time_series = pd.DataFrame({
    'Date': dates,
    'Temperature (°C)': daily_temperatures
})

# Display the first and last few rows to verify
print("First 5 rows of the time series data:")
display(df_time_series.head())

print("\nLast 5 rows of the time series data:")
display(df_time_series.tail())

print("\nDataFrame Info:")
df_time_series.info()
```

```python
# Calculate basic statistics
mean_temp = df_time_series['Temperature (°C)'].mean()
max_temp = df_time_series['Temperature (°C)'].max()
min_temp = df_time_series['Temperature (°C)'].min()
std_temp = df_time_series['Temperature (°C)'].std()

print(f"Mean Temperature: {mean_temp:.2f} °C")
print(f"Maximum Temperature: {max_temp:.2f} °C")
print(f"Minimum Temperature: {min_temp:.2f} °C")
print(f"Standard Deviation of Temperature: {std_temp:.2f} °C")
```

```python
# Plot the daily temperatures over the year
plt.figure(figsize=(14, 7))
plt.plot(df_time_series['Date'], df_time_series['Temperature (°C)'], label='Daily Temperature', color='skyblue')
plt.title('Daily Temperature Over One Year')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.tight_layout()
plt.show()

# Plot a histogram of temperatures to see the distribution
plt.figure(figsize=(10, 6))
plt.hist(df_time_series['Temperature (°C)'], bins=20, color='lightcoral', edgecolor='black', alpha=0.7)
plt.title('Distribution of Daily Temperatures')
plt.xlabel('Temperature (°C)')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

## Daily Stock Price Simulation and Analysis

### 1. Simulate Daily Stock Prices

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Set a random seed for reproducibility
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
np.random.seed(42)

# Define parameters for simulation
start_date = '2023-01-01'
end_date = '2023-12-31'
dates = pd.date_range(start=start_date, end=end_date, freq='B') # 'B' for business day frequency

# Number of data points
num_days = len(dates)

# Simulate daily returns (assume normally distributed with mean 0 and std 0.01)
daily_returns = np.random.normal(0, 0.01, num_days)

# Simulate initial stock price
initial_price = 100

# Calculate cumulative returns and stock prices
stock_prices = initial_price * np.exp(np.cumsum(daily_returns))

# Create a Pandas DataFrame
df_stock = pd.DataFrame({
    'Date': dates,
    'Price': stock_prices
})

# Set 'Date' as index
df_stock.set_index('Date', inplace=True)

print("First 5 rows of simulated stock data:")
display(df_stock.head())

print("\nLast 5 rows of simulated stock data:")
display(df_stock.tail())

print("\nDataFrame Info:")
df_stock.info()
```

```python
# Calculate daily returns
df_stock['Daily Return'] = df_stock['Price'].pct_change()

# Calculate moving averages
df_stock['20-Day MA'] = df_stock['Price'].rolling(window=20).mean()
df_stock['50-Day MA'] = df_stock['Price'].rolling(window=50).mean()

print("Stock data with Daily Returns and Moving Averages:")
display(df_stock.head())

print("\nSummary Statistics for Price, Daily Return, and Moving Averages:")
display(df_stock[['Price', 'Daily Return', '20-Day MA', '50-Day MA']].describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(14, 7))
plt.plot(df_stock.index, df_stock['Price'], label='Stock Price', color='blue', alpha=0.7)
plt.plot(df_stock.index, df_stock['20-Day MA'], label='20-Day Moving Average', color='orange', linestyle='--')
plt.plot(df_stock.index, df_stock['50-Day MA'], label='50-Day Moving Average', color='green', linestyle='-.')

plt.title('Daily Stock Prices with Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Plot daily returns distribution
plt.figure(figsize=(10, 6))
sns.histplot(df_stock['Daily Return'].dropna(), bins=50, kde=True, color='purple')
plt.title('Distribution of Daily Returns')
plt.xlabel('Daily Return')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
